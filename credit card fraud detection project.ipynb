```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("creditcard.csv")
df
```

```python
df.isnull().sum()
```

```python
non_fraud = len(df[df.Class == 0])
fraud = len(df[df.Class == 1])
print("Number of Genuine transactions: ", non_fraud)
print("Number of Fraud transactions: ", fraud)
```

```python
labels = ["Genuine", "Fraud"]
x= df.value_counts(df['Class'], sort= True)
x.plot(kind = "bar")
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(df.corr())
plt.show()
```

```python
features=df.iloc[:,:-1]
target=df.iloc[:,-1]
```

```python
features.columns
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(features,target,test_size=0.2,random_state=42)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(a)
print(cm)
print(cr)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros= RandomOverSampler(random_state=1)
xsample,ysample=ros.fit_resample(xtrain,ytrain)
ysample.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=70,max_depth=15)
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(a)
print(cm)
print(cr)
```

```python
#The Random Forest model works better than Decision Tree.  
#If we observe our dataset suffers a serious problem of class imbalance. 
#The genuine (not fraud) transactions are more than 99% with the credit card fraud transactions constituting 0.17%.
```
